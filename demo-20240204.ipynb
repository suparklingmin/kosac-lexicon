```python
from kosac.tokenizers import KomoranTokenizer, HuggingTokenizer
```

```python
sentence = '잠이 옵니다.'
```

```python
tokenizer1 = KomoranTokenizer(userdic='./tagger/user_dictionary.txt')
tokenizer1.tokenize(sentence)
```

```python
tokenizer2 = HuggingTokenizer()
tokenizer2.tokenize(sentence)
tokenizer2.tagger_type
```

```python
tokenizer3 = HuggingTokenizer(model_name='snunlp/KR-Medium')
tokenizer3.tokenize(sentence)
tokenizer3.tagger_type
```
